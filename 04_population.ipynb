```python
animIDs=['0239','2019','groot'];
animNms=['Monkey 1','Monkey 2','Monkey 3','chance level'];
redrat = 3; fs=6*redrat
fig=plt.figure(figsize=[1.8*redrat,1.5*redrat]);
for i, animID in enumerate(animIDs):
    dataPath = outputdir+'/contrast_coef_'+animID+'.pkl'
    dat = load_dictionary_pickle(dataPath)

    plt.subplot(1,3,i+1)
    slope_itm = dat['item']; slope_rnk = dat['rank']

    print(animID)
    print("=== 统计检验结果 ===")
    
    # 1. 分别与零比较 (One-sample t-test)
    t_itm, p_itm = stats.ttest_1samp(slope_itm, popmean=0)
    t_rnk, p_rnk = stats.ttest_1samp(slope_rnk, popmean=0)
    
    print(f"Entry (item) vs 0 : t = {t_itm:.3f}, p = {p_itm:.4f}  (mean = {slope_itm.mean():.4f})")
    print(f"Rank         vs 0 : t = {t_rnk:.3f}, p = {p_rnk:.4f}  (mean = {slope_rnk.mean():.4f})")
    
    # 2. 两个分布之间的差异 (Paired t-test)
    t_diff, p_diff = stats.ttest_rel(slope_itm, slope_rnk)
    print(f"Entry vs Rank     : t = {t_diff:.3f}, p = {p_diff:.4f}")
    
    # 3. 非参数检验（更稳健，推荐同时报告）
    # Wilcoxon signed-rank test (配对非参数)
    wilcox = stats.wilcoxon(slope_itm, slope_rnk)
    print(f"Wilcoxon signed-rank: statistic = {wilcox.statistic:.1f}, p = {wilcox.pvalue:.4f}")
    
    # Mann-Whitney U (如果想看独立样本效果)
    mw = stats.mannwhitneyu(slope_itm, slope_rnk, alternative='two-sided')
    print(f"Mann-Whitney U      : p = {mw.pvalue:.4f}")

    # ====================== 摘要统计 ======================
    print("\n=== 描述统计 ===")
    print(f"Entry : mean = {slope_itm.mean():.4f} ± {slope_itm.std():.4f}  "
          f"[{np.percentile(slope_itm, 2.5):.3f}, {np.percentile(slope_itm, 97.5):.3f}]")
    print(f"Rank  : mean = {slope_rnk.mean():.4f} ± {slope_rnk.std():.4f}  "
          f"[{np.percentile(slope_rnk, 2.5):.3f}, {np.percentile(slope_rnk, 97.5):.3f}]")
    # ====================== 增强版分布图 ======================

    
    # redrat = 3; fs=6*redrat
    # fig=plt.figure(figsize=[2*redrat,1.5*redrat]);
    
    plt.hist(slope_itm,bins=15,color='k',alpha=0.5,label="entry",density=True)
    plt.hist(slope_rnk,bins=15,color='gray',alpha=0.5,label="rank",density=True)
    
    # plt.axvline(0,c='gray',ls=':')
    plt.axvline(0,ymin=0,ymax=0.8, color='gray', ls='--', lw=1, alpha=0.8)  # Zero reference
    
    plt.yticks(fontsize=fs);plt.xticks(fontsize=fs)
    
    # plt.xlim(-0.3,0.3);plt.xticks([-0.25,0,0.25])
    plt.ylim(0,7);
    if i==0: 
    #     # plt.ylabel('distribution density',fontsize=fs)
        plt.yticks([])
        plt.gca().spines['left'].set_visible(False)
    else:
        plt.yticks([])
        plt.gca().spines['left'].set_visible(False)
    
    plt.xlim(-1,1);plt.xticks([-1,0,1])
    if i==0:
        plt.text(x=-0.2,y=6.2,s='****',fontsize=fs)
    elif i==1:
        plt.text(x=-0.12,y=6.2,s='****',fontsize=fs)
        plt.legend(
            frameon=False,
            fontsize=fs-4,
            loc='upper center',
            bbox_to_anchor=(0.1, 0.9),
            handlelength=1.2,          # ← 關鍵：縮小色塊長度（預設 2.0）
            handleheight=0.2,          # 可選：稍微調整高度（預設 0.7 通常不用動）
            handletextpad=0.4          # 可選：色塊與文字間距（預設 0.8），可再拉近一點
        )
        # plt.legend(frameon=False,fontsize=fs-4,loc='upper center', bbox_to_anchor=(0.1, 0.9))
    elif i==2:
        plt.text(x=-0.22,y=6.2,s='****',fontsize=fs)
        

    plt.title(animNms[i],fontsize=fs-2)
    plt.gca().spines['top'].set_visible(False);plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['bottom'].set_visible(True);
    print('')

plt.suptitle('pearsonr coefficient',x=0.58,y=0.03,fontsize=fs)
plt.tight_layout()

fig.savefig(figdir+'/'+'fig2f.pdf', format='pdf', dpi=300, bbox_inches='tight')
```
